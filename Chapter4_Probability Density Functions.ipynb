## Jensen's Inequality

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
X = np.random.uniform(0.5, 2, 1000)  # Ensuring a wider range for stronger illustration

# Convex Function: g(X) = e^X
g_convex = np.exp(X)
E_X = np.mean(X)
g_E_X = np.exp(E_X)
E_g_X = np.mean(g_convex)

plt.figure(figsize=(12, 6))
plt.plot(X, g_convex, '.', alpha=0.5, label='$g(X) = e^X$')
plt.axhline(g_E_X, color='red', linestyle='--', label='$g(E[X])$', linewidth=2)
plt.axhline(E_g_X, color='blue', linestyle='--', label='$E[g(X)]$', linewidth=2)
plt.title("Jensen's Inequality - Convex Function")
plt.xlabel('X')
plt.ylabel('$g(X)$')
plt.legend()
plt.savefig("Chapter4_Jensen_Inequality_Convex.png")
plt.show()

# Concave Function: g(X) = \log(X)
g_concave = np.log(X)
g_E_X_concave = np.log(E_X)
E_g_X_concave = np.mean(g_concave)

plt.figure(figsize=(12, 6))
plt.plot(X, g_concave, '.', alpha=0.5, label='$g(X) = \log(X)$')
plt.axhline(g_E_X_concave, color='red', linestyle='--', label='$g(E[X])$', linewidth=2)
plt.axhline(E_g_X_concave, color='blue', linestyle='--', label='$E[g(X)]$', linewidth=2)
plt.title("Jensen's Inequality - Concave Function")
plt.xlabel('X')
plt.ylabel('$g(X)$')
plt.legend()
plt.savefig("Chapter4_Jensen_Inequality_Concave.png")
plt.show()
```

## Markov's Inequality 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
data = np.random.exponential(scale=1, size=1000)

# Calculate the mean of the data
mean_data = np.mean(data)

# Define a range of 'a' values
a_values = np.linspace(1, 10, 100)

# Calculate the probability P(X >= a) from the data
empirical_prob = [np.mean(data >= a) for a in a_values]

# Calculate the bound given by Markov's inequality
markov_bound = [mean_data / a for a in a_values]

# Plot the empirical probabilities and the Markov's inequality bound
plt.figure(figsize=(12, 6))
plt.plot(a_values, empirical_prob, label='Empirical P(X >= a)', color='blue')
plt.plot(a_values, markov_bound, label='Markov\'s Inequality Bound', color='red', linestyle='--')
plt.title('Markov\'s Inequality Bound for Different Values of a')
plt.xlabel('a')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

## Chebyshev's inequality

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data from a normal distribution
np.random.seed(42)
mu, sigma = 0, 1  # mean and standard deviation
data = np.random.normal(mu, sigma, 10000)

# Define range of k values
k_values = np.linspace(1, 4, 10)

# Calculate empirical probabilities P(|X - mu| >= k * sigma)
empirical_probs = [np.mean(np.abs(data - mu) >= k * sigma) for k in k_values]

# Calculate bound given by Chebyshev's inequality
chebyshev_bounds = [1 / k**2 for k in k_values]

# Plotting normal distribution with areas highlighting |X - mu| >= k * sigma
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Normal Distribution')
for k in [1, 2, 3]:
    plt.fill_between(x, y, where=np.abs(x - mu) >= k * sigma, alpha=0.2, label=f'|X - μ| ≥ {k}σ')
plt.title('Normal Distribution with Highlighted Areas')
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the empirical probabilities and Chebyshev's bounds
plt.figure(figsize=(12, 6))
plt.plot(k_values, empirical_probs, label='Empirical Probability', marker='o', color='blue')
plt.plot(k_values, chebyshev_bounds, label='Chebyshev\'s Bound', marker='x', color='red', linestyle='--')
plt.title('Chebyshev\'s Inequality Bound Comparison')
plt.xlabel('k')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

## PDF Transformation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.signal import convolve

# 1. PDF Addition (Convolution)
# Define two normal distributions
mu1, sigma1 = 0, 1  # mean and standard deviation for the first distribution
mu2, sigma2 = 2, 1.5  # mean and standard deviation for the second distribution

# Generate a range of x values
x = np.linspace(-10, 10, 1000)
pdf1 = norm.pdf(x, mu1, sigma1)
pdf2 = norm.pdf(x, mu2, sigma2)
pdf3 = norm.pdf(x, mu1+mu2, 1.8)

# Compute the convolution of the two PDFs
pdf_convolution = convolve(pdf1, pdf2, mode='same') * (x[1] - x[0])

# Plot the original PDFs and the resulting convolved PDF
plt.figure(figsize=(12, 6))
plt.plot(x, pdf1, label='PDF of N(0, 1)')
plt.plot(x, pdf2, label='PDF of N(2, 1.5)')
plt.plot(x, pdf3, label='PDF of N(2, 1.8)')
plt.plot(x, pdf_convolution, label='Convolution of PDFs', linestyle='--', linewidth=2)
plt.title('PDF Addition (Convolution)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.savefig("Chapter4_PDF_Addition.png")
plt.show()
```

## Maximum of r.v.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to compute the CDF of the maximum
def cdf_max(cdf_x, N):
    return cdf_x ** N

# Function to compute the PDF of the maximum
def pdf_max(cdf_x, pdf_x, N):
    return N * (cdf_x ** (N - 1)) * pdf_x

# Generate data for a normal distribution
mu, sigma = 0, 1  # mean and standard deviation
x = np.linspace(-3, 3, 1000)
cdf_x = norm.cdf(x, mu, sigma)
pdf_x = norm.pdf(x, mu, sigma)

# Define different values of N
N_values = [1, 2, 5, 10, 20]

# Plot the CDF of the maximum for different values of N
plt.figure(figsize=(11, 9))
for N in N_values:
    cdf_y = cdf_max(cdf_x, N)
    plt.plot(x, cdf_y, label=f'N={N}')
plt.title('CDF of Maximum for Different Values of N')
plt.xlabel('y')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.savefig("Chapter4_Maximum_of_rv_CDF.png")
plt.show()

# Plot the PDF of the maximum for different values of N
plt.figure(figsize=(11, 9))
for N in N_values:
    pdf_y = pdf_max(cdf_x, pdf_x, N)
    plt.plot(x, pdf_y, label=f'N={N}')
plt.title('PDF of Maximum for Different Values of N')
plt.xlabel('y')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)
plt.savefig("Chapter4_Maximum_of_rv_PDF.png")
plt.show()
```

## Random process, Brownian motion

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate Brownian motion paths
def simulate_brownian_motion(T, N, mu, sigma, num_paths):
    dt = T / N
    t = np.linspace(0, T, N)
    paths = np.zeros((num_paths, N))
    
    for i in range(num_paths):
        W = np.zeros(N)
        W[1:] = np.cumsum(np.random.normal(0, np.sqrt(dt), N-1))
        paths[i] = mu * t + sigma * W
    
    return t, paths

# Parameters
T = 10
N = 10000
mu = 0.0
sigma = 0.2
num_paths = 100

# Simulate multiple Brownian motion paths
t, paths = simulate_brownian_motion(T, N, mu, sigma, num_paths)

# Calculate the ensemble mean
ensemble_mean = np.mean(paths, axis=0)

# Calculate the time-mean for each path
time_means = np.mean(paths, axis=1)

# Calculate the average of time-means
average_time_mean = np.mean(time_means)

# Plot the sample paths
plt.figure(figsize=(12, 6))
for i in range(num_paths):
    plt.plot(t, paths[i], color='blue', alpha=0.1)

plt.plot(t, ensemble_mean, color='red', label='Ensemble Mean', linewidth=2)
plt.axhline(average_time_mean, color='green', linestyle='--', label='Average Time-Mean', linewidth=2)
plt.title('Brownian Motion: Sample Paths, Ensemble Mean, and Average Time-Mean')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.savefig("Chapter4_Brownian_Motion_1.png")
plt.show()
```

## Markov process

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Define the transition matrix and states
states = ['Sunny', 'Cloudy', 'Rainy']
transition_matrix = [
    [0.8, 0.15, 0.05],  # Sunny
    [0.2, 0.6, 0.2],    # Cloudy
    [0.1, 0.3, 0.6]     # Rainy
]

# Function to plot the state transition diagram
def plot_markov_chain(transition_matrix, states):
    G = nx.DiGraph()
    for i, state in enumerate(states):
        for j, prob in enumerate(transition_matrix[i]):
            if prob > 0:
                G.add_edge(state, states[j], weight=prob)
    
    pos = nx.spring_layout(G)
    edge_labels = {(states[i], states[j]): f'{prob:.2f}' for i, row in enumerate(transition_matrix) for j, prob in enumerate(row) if prob > 0}
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=15, font_weight='bold')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
    plt.title('Discrete-Time Markov Chain: Weather States')
    plt.savefig("Chapter4_Markov_1.png")
    plt.show()

plot_markov_chain(transition_matrix, states)

# Simulate a sequence of states based on the Markov chain
def simulate_markov_chain(transition_matrix, states, initial_state, num_steps):
    current_state = initial_state
    state_sequence = [current_state]
    
    for _ in range(num_steps):
        current_state = np.random.choice(states, p=transition_matrix[states.index(current_state)])
        state_sequence.append(current_state)
    
    return state_sequence

initial_state = 'Sunny'
num_steps = 100
state_sequence = simulate_markov_chain(transition_matrix, states, initial_state, num_steps)

# Plot the state sequence over time
def plot_state_sequence(state_sequence, states):
    state_indices = [states.index(state) for state in state_sequence]
    time_steps = np.arange(len(state_sequence))

    plt.figure(figsize=(12, 6))
    plt.step(time_steps, state_indices, where='mid', linewidth=2)
    plt.yticks(range(len(states)), states)
    plt.title('Markov Chain State Sequence Over Time')
    plt.xlabel('Time Steps')
    plt.ylabel('States')
    plt.grid(True)
    plt.savefig("Chapter4_Markov_2.png")
    plt.show()

plot_state_sequence(state_sequence, states)
```
