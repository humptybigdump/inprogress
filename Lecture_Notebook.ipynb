```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sequence of numbers
x = np.arange(1, 101)
# Add noise to the data
noise = np.random.normal(0, 1, 100)
x = x + noise

# Generate data with an exponential trend
y = np.exp(x / 20)

# Plot the data
plt.plot(x, y, label='y=exp((x+noise)/20)', color='blue')
plt.title('Data with an exponential trend')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.xlim(0, 100)
plt.show()
```

```python
# Take the natural logarithm of the data
y2 = np.log(y)

# Plot the data
plt.plot(x, y2, label='y=(x+noise)/20')
plt.legend()
plt.title('Logged data has a linear trend')
plt.xlabel('x')
plt.show()
```

```python
y3 = np.log(y[1:]) - np.log(y[:-1])

# Plot the data
plt.plot(x[1:], y3)
plt.title('Logged diffs have no trend')
plt.show()
```

```python
# Check the shape of the arrays
print(f"Shape of x {np.shape(x)}")
print(f"Shape of y3: {np.shape(y3)}")  
print(f"Shape of x[1:]: {np.shape(x[1:])}")
```

```python
# Check the shape of the arrays utilizing that the numpy object have a shape attribute
print(f"Shape of x {x.shape}")
print(f"Shape of y3: {y3.shape}")  
print(f"Shape of x[1:]: {x[1:].shape}")
```
