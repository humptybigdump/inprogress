```python
import numpy as np
from sklearn.metrics import (mean_absolute_error,
                             mean_squared_error)

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
```

```python
# Import the required packages
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
# Import the data
titanic = pd.read_csv("data/titanic_clean.csv")
# Separate X and y
X = titanic.drop("Survived", axis=1)
y = titanic["Survived"]
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Random 80% sample split
X_train, X_test, y_train, y_test = (
    train_test_split(X_scaled, y,
                     test_size = 0.2,
                     shuffle = True,
                     random_state = 1234))
# Estimate the model
clf = KNeighborsClassifier(
    n_neighbors=10).fit(X_train, y_train)
# Predict the test set
y_pred = clf.predict(X_test)
# Accuracy
accuracy_score(y_test, y_pred)
```

## Regression

```python
# Import the required packages
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Import the data
mel = pd.read_csv("data/Mel_housing_clean.csv")

# Remove 1st two character columns
mel = mel.iloc[:, 2:]

# Separate X and y
X = mel.drop("Price", axis=1)
y = mel["Price"]

# Random 80% sample split
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size = 0.2,
                     shuffle = True, random_state = 1234)
```

```python
# Train the model
mod_fit = LinearRegression().fit(X_train, y_train)
# Predict on the test data
predictions = mod_fit.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.title("Actual vs. Predicted Prices")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='black')
plt.show()
```
